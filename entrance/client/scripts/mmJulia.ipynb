```julia
bound = (-8.7015, -8.5302, 40.0990, 41.2082)
```

```julia
function inregion(trip::Trip, bound::Tuple)
    (min_lon, max_lon, min_lat, max_lat) = bound
    function inregion(lon, lat)
        min_lon <= lon <= max_lon &&
        min_lat <= lat <= max_lat
    end
    all(inregion.(trip.lon, trip.lat))
end
```
